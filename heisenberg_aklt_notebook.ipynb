```julia
using TensorKit
```

```julia
V = SU2Space(1=>1)
SpinSpin = TensorMap(zeros,V⊗V,V⊗V)
Casimir(s::SU2Irrep) = s.j *(s.j+1)
for (f1,f2) in fusiontrees(SpinSpin)
    SpinSpin[f1,f2] .= 0.5*(Casimir(f1.coupled) - Casimir(f2.uncoupled[1]) - Casimir(f2.uncoupled[2]))
end
```

```julia
W = fuse(V⊗V)
```

```julia
CG = isomorphism(W,V⊗V);
SpinSpinFused = CG*SpinSpin*adjoint(CG)
```

```julia
function HaldaneModel(beta,nsites)
    V = SU2Space(1=>1)
    SpinSpin = TensorMap(zeros,V⊗V,V⊗V)
    Casimir(s::SU2Irrep) = s.j *(s.j+1)
    for (f1,f2) in fusiontrees(SpinSpin)
        SpinSpin[f1,f2] .= 0.5*(Casimir(f1.coupled) - Casimir(f2.uncoupled[1]) - Casimir(f2.uncoupled[2]))
    end
    H = SpinSpin + beta*SpinSpin*SpinSpin
end
```
